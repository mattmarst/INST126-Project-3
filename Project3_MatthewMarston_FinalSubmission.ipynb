```python
logs = {}

log_list = text.split()
# print(log_list)
log_length = len(log_list)


    
```

```python
# log dictionary creation
log_num = int(log_length/5)
count = 0

for i in range (log_num):
    logs[i] = {'Date' : log_list[0 + count], 'Time' : log_list[1 + count], 'Activity' : log_list[2 + count], 'Server' : log_list[3 + count], 'User Email-ID' : log_list[4 + count]}
    count += 5


#for log in logs:
#    print(logs[log])
    
    
```

```python
# unique users list

def uniqueUsers(logs):
    users = {"remove"}
    users.remove("remove")
    
    for log in logs:
        users.add(logs[log]['User Email-ID'])
        
    users = list(users)
    
    return users
```

```python
# suspicious report

def suspiciousReport(logs):
    
    users = uniqueUsers(logs)
    sus_count = 0
    sus_dict = {}
    confirmed_sus = 0
    
    for user in users:
        for day in range(31):
            for log in logs:
                if ((logs[log]['Date'] == f"2020-05-{day + 1}") and (logs[log]['User Email-ID'] == user) and (logs[log]['Activity'] == "login")):
                    sus_count += 1
                    
                    sus_dict.update({f'Log {sus_count}': {'Time': logs[log]['Time'], 'Activity': logs[log]['Activity'], 'Server': logs[log]['Server']}})
            if (sus_count >= 5):
                    print(f'Date: 2020-05-{day + 1}')
                    print(f'{user} is sus')
                    #for sus_log in sus_dict:
                    #    print(f'{sus_dict[sus_log]['Time']}, {sus_dict[sus_log]['Activity']}, {sus_dict[sus_log]['Server']}')
                    #print(f'{sus_dict[0]['Time']}, {sus_dict['Activity']}, {sus_dict['Server']}')
                    
                    
                    print(sus_dict)
                    confirmed_sus += 1
                    sus_count = 0

            else:
                sus_count = 0
```
